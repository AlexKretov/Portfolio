### Работа с пропусками

```python
#Определим, в каких столбцах есть пропуски.
has_balnks=data.isna().sum()
has_balnks=has_balnks[has_balnks!=0]
has_balnks
```

```python
### Заменим число балконов на 0 там где оно пропущено.
data['balcony']=data['balcony'].fillna(0)
data['locality_name']=data['locality_name'].fillna("не указано")
#Заменим пропуски названия локации на Санкт-Петербург, если они не дальше от центра, чем питерский максимум
max=data[data['locality_name']=='Санкт-Петербург']['cityCenters_nearest'].max()
data.loc[(data['locality_name']=="не указано") & (data['cityCenters_nearest']<=max),'locality_name']="Санкт-Петербург"
#Заменим пропуски аппартаментов на обычные квартиры
data['is_apartment']=data['is_apartment'].fillna(False)
#Напишем функцию заменющую пропуски в указанном столбце на медианные для данной локации значения:
def mediate_nans(nans_array):
    #Сначала найдём медианное значение искомого параметра nans_array для каждой локации
    parks_nearest_medians=data.pivot_table([nans_array],['locality_name'], aggfunc='median')
    #выкинем те локации, где медиану найти не удалось
    parks_nearest_medians=parks_nearest_medians.dropna()
    #созданим словарь типа "название локации":"медиана искомого параметра"
    listOfStr=list(parks_nearest_medians.index)
    listOfInt=list(parks_nearest_medians[nans_array])
    dictOfWords = dict(zip(listOfStr, listOfInt))
    #Проведём замену пропусков на значения из словаря
    data[nans_array]= data[nans_array].fillna(data['locality_name'].map(dictOfWords))

    
#теперь, применим функцию к удалённости от парков
mediate_nans('parks_nearest') 
mediate_nans('parks_around3000')# к количеству парков
mediate_nans('ponds_nearest') # к удаленности от прудов
mediate_nans('ponds_around3000')# к количеству прудов
# а вот с аэропортами эта история не пройдёт, наверное


#проверим, что число пропусков сократилось
has_balnks=data.isna().sum()
has_balnks=has_balnks[has_balnks!=0]
has_balnks
```

**Причины появления пропусков**
1. Почему-то выгрузили не те данные или не всё, что есть. Либо люди не стали выгружать, либо ПО дало сбой.
2. Ошибки в алгоритмах, считающих расстояния по карте: скажем, не смогли определить центр какого-нибудь ПГТ, или поспорили где центр Питера - в Смольном или Эрмитаже.
3. Не тот формат: например, вместо сантиметров записали метры или словами, например высота потолков: "Высокие" и система восприяла это как NaN.
4. Изначально часть данных вводилась самими продавцами, а они могли в спешке или по незнанию, не заполнить необязательные параметры.
5. В источник мог вносить фальшивые данные рекламный бот, чтобы на их фоне предложения его хозяина смотрелись более конкурентно.

### Теперь посмотрим как удалённость от центра Санкт-Петербурга влияет на цену недвижимости

Ранее вы посчитали расстояние до центра в километрах. Теперь выделите квартиры в Санкт-Петербурге с помощью столбца locality_name и вычислите среднюю цену каждого километра. Опишите, как стоимость объектов зависит от расстояния до центра города.

```python
#вынрузим данные по Питеру в отдельный датафрейм
far_from_centr=data[data['locality_name']=='Санкт-Петербург']
#сделаем сводную таблицу с интересующими показателями
far_from_centr=far_from_centr.pivot_table(['last_price'],['dist_to_centr'], aggfunc=['median'])
#отрежем хвост, где очень мало сделок и одни выбросы
far_from_centr=far_from_centr.loc[:24]
far_from_centr.columns=["средняя цена"]
far_from_centr.plot( y='средняя цена', style='o', xlim=(0, 25), grid=True, figsize=(10, 5),ylabel='средняя цена', xlabel='расстояние до центра, км', title='Соотношение цены и расстояния до центра')
plt.show()
far_from_centr.columns=["средняя цена за кв.м."]
far_from_centr=far_from_centr.dropna()
ind=list(far_from_centr.index)
#Теперь у нас есть всё, чтобы посчитать среднюю цену киллометра.
far_from_centr['средняя цена киллометра']=round(
    far_from_centr["средняя цена за кв.м."]/far_from_centr.index
)
#сортируем по удалению от центра
far_from_centr=far_from_centr.sort_values(['средняя цена киллометра'], ascending=False, inplace=False)
#убираем столбец промежуточных вычислений
far_from_centr = far_from_centr.drop(['средняя цена за кв.м.'], axis=1) 
display(far_from_centr)
```
